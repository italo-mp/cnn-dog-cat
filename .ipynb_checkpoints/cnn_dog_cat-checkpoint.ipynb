## DOG VS CAT
  Projeto utilizando CNN para identificação de cachorros ou gatos.

```python
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout
import numpy as np
from PIL import Image
```

```python
cnn_clf = Sequential()
conv1 = Conv2D(32, (3,3), input_shape = (64,64,3), activation = 'relu')
cnn_clf.add(conv1)

pool1 = MaxPooling2D(pool_size = (2,2))
cnn_clf.add(pool1)

conv2 = Conv2D(64, (3,3), activation = 'relu')
cnn_clf.add(conv2)

pool2 = MaxPooling2D(pool_size = (2,2))
cnn_clf.add(pool2)

cnn_clf.add(Flatten())
```

```python
h1 = Dense(units = 64, activation = 'relu')
cnn_clf.add(h1)
cnn_clf.add(Dropout(0.3))

h2 = Dense(units = 1, activation = 'sigmoid')
cnn_clf.add(h2)
```

```python
cnn_clf.compile(optimizer = 'adam', loss= 'binary_crossentropy', metrics = ['accuracy'])
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_setgen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)
test_setgen = ImageDataGenerator(rescale = 1./255)
```

```python
training_data = train_setgen.flow_from_directory('./animal_data/training_set', 
                                                target_size = (64,64), batch_size = 32, class_mode = 'binary')

test_data = test_setgen.flow_from_directory('./animal_data/test_set', 
                                                target_size = (64,64), batch_size = 32, class_mode = 'binary')
```

```python
cnn_clf.fit_generator(training_data, steps_per_epoch = (8000/32), epochs = 15, validation_data = test_data,
                     validation_steps = (2000/32))
```

```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
import numpy as np
import os


def load_predict_image(img_path):

    img = image.load_img(img_path, target_size=(64, 64))
    img_tensor = image.img_to_array(img)                    
    img_tensor = np.expand_dims(img_tensor, axis=0)        
    img_tensor /= 255.                                    

    plt.imshow(img_tensor[0])                           
    plt.axis('off')
    predict = cnn_clf.predict(img_tensor)

    if predict > 0.5:
        plt.title("%.2f" % (predict[0]*100) + "% dog")
    else:
        plt.title("%.2f" % ((1-predict[0])*100) + "% cat")
        
    plt.show()
```

```python
import os

path_images = './images_test/'
for file in os.listdir(path_images):
    load_predict_image(path_images+file)
```
